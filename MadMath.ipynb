```python
import pandas
import scipy.stats
with open("data/ttestdata.csv") as csv_file:
    my_file = pandas.read_csv(csv_file)
winner_y = my_file.iloc[:,0]
loser_y = my_file.iloc[:,1]
print(winner_y)
print(loser_y)

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

def make_mean_line(x_values, y_values):
    avg_x = [np.average(x_values) - .2, np.average(x_values) + .2]
    avg_y = [np.average(y_values) for i in list(range(2))]
    return avg_x, avg_y

fig = plt.figure()
fig.suptitle('NCAA Winners vs. Losers (ppg)', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
ax.set_ylabel('points per game')

winner_x = [1 for i in list(range(len(winner_y)))]
winner_avg_x, winner_avg_y = make_mean_line(winner_x, winner_y)

loser_x = [2 for i in list(range(len(loser_y)))]
loser_avg_x, loser_avg_y = make_mean_line(loser_x, loser_y)

plt.plot(winner_x, winner_y, 'ro')
plt.plot(loser_x, loser_y, 'bo')
plt.plot(winner_avg_x, winner_avg_y,'r')
plt.plot(loser_avg_x, loser_avg_y,'b')
plt.axis([0, 3, 60, 90])
frame1 = plt.gca()
frame1.axes.get_xaxis().set_visible(False)
ax.text(0.8, 58, 'Winners')
ax.text(1.85, 58, 'Losers')
plt.show()

scipy.stats.ttest_ind(winner_y, loser_y)
```

```python
#plot normal distribution of 64 teams
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
import scipy.stats

fig = plt.figure()
fig.suptitle('64 Team Probability Distribution', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)

ax.set_xlabel('points per game')
ax.set_ylabel('probability of winning')

mu = 74.14
variance = 10.93
sigma = math.sqrt(variance)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.plot(x,scipy.stats.norm.pdf(x, mu, sigma))
plt.show()

#plot normal distribution of 32 winning teams

fig = plt.figure()
fig.suptitle('Winning Team Probability Distribution', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)

ax.set_xlabel('points per game')
ax.set_ylabel('probability of winning')

mu = 68.24
variance = 5.06
sigma = math.sqrt(variance)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.plot(x, scipy.stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
# the histogram of the samples taken from the 32 winners posterior distribution, along with the probability density function:
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()
fig.suptitle('Posterior Sampling Distribution', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)

ax.set_xlabel('points per game')
ax.set_ylabel('probability of winning')

np.random.normal(loc=68.24, scale=5.06, size=None)
s = np.random.normal(mu, sigma, 1000)

count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
